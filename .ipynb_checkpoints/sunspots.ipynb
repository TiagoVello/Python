```python
import warnings
import urllib3
import matplotlib.pyplot as plt
import pandas as pd
import torch
import numpy as np
from statistics import mean
from scipy.stats import pearsonr
from numpy.linalg import inv
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
http = urllib3.PoolManager()
r = http.request('GET', 'https://www.esrl.noaa.gov/psd/gcos_wgsp/Timeseries/Data/sunspot.long.data') 
sunspot = str(r.data)
sunspot = sunspot[14:-448]
sunspot = sunspot.replace('\\', '  ')
sunspot = sunspot.split('  ')
new_list = []
errors = []
for i in range(len(sunspot)):
    try:
        new_list.append(float(sunspot[i]))
    except ValueError:
        errors.append(sunspot[i])
sunspot = new_list 
plt.plot(list(range(len(sunspot))), sunspot)
plt.title('Sunspots by months')
plt.show
```

```python
list_test = sunspot[-5*12:]
list_train = sunspot[:-5*12]
print(len(list_train))
print(len(list_test))
```

```python
x_test = np.ndarray(shape=(40,20),dtype=float)
x_train = np.ndarray(shape=(3163,20),dtype=float)
y_test = np.ndarray(shape=(40),dtype=float)
y_train = np.ndarray(shape=(3163),dtype=float)

for i in range(len(list_train)-20):
    for j in range(21):
        if j != 20:
            x_train[i, j] = list_train[i+j]
        elif j == 20:
            y_train[i] = list_train[i+j]

for i in range(len(list_test)-20):
    for j in range(21):
        if j != 20:
            x_test[i, j] = list_test[i+j]
        elif j == 20:
            y_test[i] = list_test[i+j]
```

```python
print(x_test.shape)
print(x_train.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
class RegressaoLinear:
    def __init__(self, bias = False):
        self.w = None
        self.x = None
        self.y = None
        self.bias = bias
    
    def fit(self, x, y, alpha = 0):
        if self.bias == True:
            b = np.ones((x.shape[0],x.shape[-1]+1))
            b[:,1:] = x
            self.x = b
        else :
            self.x = x
        self.y = y
        i = np.identity(self.x.shape[-1])
        self.w = np.dot(np.dot(inv(np.dot(self.x.T, self.x)+(alpha*i)),self.x.T),self.y)
    def coef(self):
        if self.bias == True:
            print('Coeficientes :')
            print(self.w[1:])
            print('Bias :')
            print(self.w[0])
        else :
            print('Coeficientes :')
            print(self.w)
        
    def pred(self, x):
        if self.bias == True:
            b = np.ones((x.shape[0],x.shape[-1]+1))
            b[:,1:] = x
            return np.dot(b, self.w.T)
        else:
            return np.dot(x, self.w.T)
    
    def rmse(self, x, y):
        if self.bias == True:
            b = np.ones((x.shape[0],x.shape[-1]+1))
            b[:,1:] = x
            return np.sum(np.sqrt((y - np.dot(b, self.w.T))**2))/len(y)
        else: 
            return np.sum(np.sqrt((y - np.dot(x, self.w.T))**2))/len(y)
        
```

```python
l = RegressaoLinear(bias = True)
l.fit(x_train,y_train)
l.coef()
print('RMSE do conjunto de treinamento: {}'.format(l.rmse(x_train, y_train)))
print('RMSE do conjunto de teste: {}'.format(l.rmse(x_test, y_test)))
```

```python
# Utilizando a estratégia wrapper considerando k atrasos 

def dataset (list_train, k):
    x = np.ndarray(shape=(len(list_train)-k,k),dtype=float)
    y = np.ndarray(shape=(len(list_train)-k),dtype=float)
    for i in range(len(list_train)-k):
        for j in range(k+1):
            if j != k:
                x[i, j] = list_train[i+j]
            elif j == k:
                y[i] = list_train[i+j]
    
    return x, y
            
def k_fold (x, y, alpha, k = 10):
    erro_total = 0
    for fold in range(k):
        x_mask = np.full(x.shape[0], True)
        x_mask[int((fold/10)*x.shape[0]):int(((fold+1)/10)*y.shape[0])] = False
        y_mask = np.full(y.shape[0], True)
        y_mask[int((fold/10)*y.shape[0]):int(((fold+1)/10)*y.shape[0])] = False
        x_train = x[x_mask]
        y_train = y[y_mask]
        x_test = x[~x_mask]
        y_test = y[~y_mask]
        modelo_linear = RegressaoLinear(bias = True)
        modelo_linear.fit(x_train, y_train, alpha = alpha)
        erro_total += modelo_linear.rmse(x_test, y_test)
        
    return erro_total/k
```

```python
estatisticas = {'RMSE':[],'K':[],'Alpha':[]}

alpha_range = []
for i in range(20):
    alpha_range.append(1100 + 10*i)
print(alpha_range)
    
for k in range(40):
    for alpha in alpha_range:
        x, y = dataset(list_train, k)
        estatisticas['RMSE'].append(k_fold(x, y, alpha))
        estatisticas['K'].append(k)
        estatisticas['Alpha'].append(alpha)
```

```python
df = pd.DataFrame(estatisticas)
df.sort_values(by='RMSE').head(10)
```

```python
x, y = dataset(list_train, 40)
pear = []
for i in range(40):
    pear.append(pearsonr(x[:,39-i], y)[0])
plt.plot(range(len(pear)), pear)
plt.show
```

Podemos observar, pelos coeficientes de pearson, que as observações mais recentes são as mais informativas no desafio de prever
a próxima mancha solar. Como a correlação decresce de forma consistente, é de se esperar que as primeiras k observações sejam 
as mais úteis. O número "ótimo" de k achado foi de 27 observações para um alpha de regularização de 1230. O RMSE achado com 
esses parametros foi de 17,67.

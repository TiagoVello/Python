```python
import random
import numpy
import matplotlib.pyplot as plt
```

```python
# Inicialização
n = 2000
portas = {'A','B','C'}
```

```python
resultado_quando_trocou = [0]*(n+1)
resultado_quando_nao_trocou = [0]*(n+1)
for i in range(n):
    # Escolha aleatória da porta que contem o prêmio
    porta_premio = random.choice(list(portas))
    # Escolha do participante
    porta_escolhida = random.choice(list(portas))
    # Apresentador mostra uma porta livre
    porta_aberta = random.choice(list(portas.difference(porta_premio).difference(porta_escolhida))) 
    # Porta escolhida após a troca
    porta_trocada = list(portas.difference(porta_escolhida).difference(porta_aberta))[0]
    # Contagem dos resultados
    if porta_escolhida == porta_premio:
        resultado_quando_nao_trocou[i] = 1
        resultado_quando_trocou[i] = 0
    if porta_trocada == porta_premio:
        resultado_quando_nao_trocou[i] = 0
        resultado_quando_trocou[i] = 1

print('{} % de acerto quando o participante troca de porta'.format(numpy.mean(resultado_quando_trocou)*100))
print('{} % de acerto quando o participante nao troca de porta'.format(numpy.mean(resultado_quando_nao_trocou)*100))
```

```python
for i in range(n):
    resultado_quando_trocou[i+1] = resultado_quando_trocou[i+1] + resultado_quando_trocou[i]
    resultado_quando_nao_trocou[i+1] = resultado_quando_nao_trocou[i+1] + resultado_quando_nao_trocou[i]
for i in range(n):
    resultado_quando_trocou[i+1] = resultado_quando_trocou[i+1]/(i+1)
    resultado_quando_nao_trocou[i+1] = resultado_quando_nao_trocou[i+1]/(i+1)
    
```

```python
fig, axs = plt.subplots()

# We can set the number of bins with the `bins` kwarg
axs.plot(list(range(n+1)), resultado_quando_trocou)
axs.set(xlabel = 'Numero da simulação', ylabel = 'Probabilidade', title = 'Probabilidade de acerto quando o participante troca de porta')
```

```python
fig, axs = plt.subplots()
axs.plot(list(range(n+1)), resultado_quando_nao_trocou)
axs.set(xlabel = 'Numero da simulação', ylabel = 'Probabilidade', title = 'Probabilidade de acerto quando o participante não troca de porta')
```

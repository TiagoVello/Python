```python
from random import choice
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
population = [1]*3000 + [0]*7000
```

```python
trial_a = []
trial_b = []
trial_range = list(range(2000))

# Experimento A
na = 10
for _ in trial_range:
    trial = []
    for _ in range(na):
        trial.append(choice(population))
    trial_a.append(sum(trial)/na)
    
# Experimento B
nb = 500
for _ in trial_range:
    trial = []
    for _ in range(nb):
        trial.append(choice(population))
    trial_b.append(sum(trial)/nb)
```

```python
plt.plot(trial_range,trial_a,'b',trial_range,trial_b,'r')
```

Para o experimento A (representado em azul), foram retirados 10 amostras randomicamentes selecionadas de uma população (com reposição) que contem 1 (representando o candidato A) e 0 (representando o candidato B) com p = 0.3. Então foi somado as 10 amostras e se normaliza pelo numero de amostras retiradas para que se obtenha a proporção de votos para o candidato A. O experimento foi repetido 2000 vezes e plotado no gráfico acima. Para o experimento B, (em vermelho) foram retiradas 500 amostras da população em vez de apenas 10. Podemos observar que quanto mais amostras retiramos, menor é a variância, mas a média continua em torno da proporção da população.

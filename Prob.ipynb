```python
from random import choice
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
population = [1,0]
```

```python
trial_a = []
trial_b = []
trial_range = list(range(100))

# Trial A
for _ in trial_range:
    trial = []
    for _ in range(10):
        trial.append(choice(population))
    trial_a.append(sum(trial)/10)
    
# Trial B
for _ in trial_range:
    trial = []
    for _ in range(500):
        trial.append(choice(population))
    trial_b.append(sum(trial)/500)
```

```python
plt.plot(trial_range,trial_a,'b',trial_range,trial_b,'r')
```

Para o teste A (representado em azul), foram retirados 10 amostras randomicamentes selecionadas de uma população que contem 0 e 1 em um proporção de 1:1. então foi somado as 10 amostras e se normaliza pelo numero de amostras retiradas para que se obtenha a proporção de amostras 1 retiradas. O experimento foi repetido 100 vezes e plotado no gráfico acima. Para o teste B, (em vermelho) foram retiradas 100 amostras da população em vez de apenas 10. Podemos observar que quanto mais amostras retiramos, menor é a variância, mas a média continua em torno da proporção da população.
